```python
# data
import numpy as np
import pandas as pd

restaurants = pd.read_csv("https://raw.githubusercontent.com/xoyeon/PDPTW/main/grubhub/0o100t100s1p100/restaurants.txt", sep = '\t').to_numpy()
orders = pd.read_csv("https://raw.githubusercontent.com/xoyeon/PDPTW/main/grubhub/0o100t100s1p100/orders.txt", sep = '\t').to_numpy()
couriers = pd.read_csv("https://raw.githubusercontent.com/xoyeon/PDPTW/main/grubhub/0o100t100s1p100/couriers.txt", sep = '\t').to_numpy()

speed = 50
costPerKilo = 2000
wage = 9160

# 1. order를 랜덤하게 발생시킴 (일단 3개)
order = orders[np.random.choice(len(orders), size = 3, replace=False)]
o1,o2,o3 = order
print(order)

# restaurant 찾기
restaurant = []
for i in range(len(order)):
    r = restaurants[restaurants[:,0] == order[i][4]]
    restaurant.append(r)
r1,r2,r3 = restaurant
print(restaurant)

# 2. Order의 'placement_time'과 'ready_time'에 근무중인 courier('on_time','off_time') 중에서
courier = []
for i in range(len(order)):
    c = couriers[(couriers[:,3] <= order[i][3]) & (couriers[:,4] >= order[i][5])]
    courier.append(c)
c1,c2,c3 = courier
print(courier)
```
