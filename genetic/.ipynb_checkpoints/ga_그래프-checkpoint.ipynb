```python
import matplotlib.pyplot as plt
import numpy as np

# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25

# 연도가 4개이므로 0, 1, 2, 3 위치를 기준으로 삼음
index = np.arange(8)

# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(index, t1, bar_width, label='no genetic')

b2 = plt.bar(index + bar_width, t2, bar_width, label='genetic(distance)')

b3 = plt.bar(index + 2 * bar_width, t3, bar_width,label='genetic(proposed)')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 8 + bar_width, 1), generation)

# x축, y축 이름 및 범례 설정
plt.xlabel('Generation', size = 13)
plt.ylabel('Trustiness', size = 13)
plt.legend()
plt.show()
```
