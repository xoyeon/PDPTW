```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = ["0.3", "0.5", "0.7"]
std = [0.1291, 0.0441, 0.0440]
trust = [0.0301, 0.2214, 0.0401]
```

```python
# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(8,5))
bar_width = 0.25

# 연도가 4개이므로 0, 1, 2, 3 위치를 기준으로 삼음
index = np.arange(3)

# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(index, std, bar_width, label='std')

b2 = plt.bar(index + bar_width, trust, bar_width, label='trust')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 3 + bar_width, 1), alpha)

# x축, y축 이름 및 범례 설정
plt.xlabel('alpha', size = 13)
plt.ylabel('value', size = 13)
plt.title("NOT DINNER")
plt.legend()

plt.grid(True, axis='y')
plt.show()
```
