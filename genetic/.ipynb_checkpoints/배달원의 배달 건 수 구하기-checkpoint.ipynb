```python
#library
import numpy as np
import pandas as pd

# data
restaurants = pd.read_csv("https://raw.githubusercontent.com/xoyeon/PDPTW/main/grubhub/0o100t100s1p100/restaurants.txt", sep = '\t')
orders = pd.read_csv("https://raw.githubusercontent.com/xoyeon/PDPTW/main/grubhub/0o100t100s1p100/orders.txt", sep = '\t')
couriers = pd.read_csv("https://raw.githubusercontent.com/xoyeon/PDPTW/main/grubhub/0o100t100s1p100/couriers.txt", sep = '\t')
rest_orders = pd.merge(restaurants, orders, how='inner', on=['restaurant'])
rest_orders.columns = ['restaurant', 'rx', 'ry', 'order', 'ox', 'oy', 'placement_time', 'ready_time']
```

```python
ideliver = []
for a in range (len(couriers)):
    o = rest_orders[(rest_orders['placement_time']>=couriers.iloc[a,3]) & (rest_orders['ready_time'] <= couriers.iloc[a,4])]
    ideliver.append(o)
print(ideliver)
```

```python
deliver_sum = 0
for b in range(len(ideliver)):
    deliver_sum += len(ideliver[b])
print(deliver_sum)
```

```python
# 배달 건 수 다 더해서 배달원 수로 나누는 평균 배달 건수
mean = deliver_sum / len(couriers)
mean
```

```python
# 배달원마다 (배달건수 - 평균)**2 을 다 더해서 나누기 배달원 수
var = 0
for c in range(len(ideliver)):
    var += ((len(ideliver[0]) - mean)**2) / len(couriers)
print(var)
```

```python
# 여기에 루트
std = var ** 0.5
print(std)
```
